### Date time preprocessing

```python
full_data['RecentTrainingExperience1To'] = full_data['RecentTrainingExperience1To']\
.mask(full_data['RecentTrainingExperience1To'] == 'PRESENT', full_data['ApplyDate'] )
full_data['RecentTrainingExperience2To'] = full_data['RecentTrainingExperience2To']\
.mask(full_data['RecentTrainingExperience2To'] == 'PRESENT', full_data['ApplyDate'] )
full_data['RecentTrainingExperience3To'] = full_data['RecentTrainingExperience3To']\
.mask(full_data['RecentTrainingExperience3To'] == 'PRESENT', full_data['ApplyDate'] )
```

```python
full_data['ApplyDate_datetime'] = pd.to_datetime(full_data['ApplyDate'], format="%Y-%m-%d")
full_data['JobPostingDate_datetime'] = pd.to_datetime(full_data['JobPostingDate'], format="%Y-%m-%d")
full_data['RecentTrainingExperience1From_datetime'] = pd.to_datetime\
(full_data['RecentTrainingExperience1From'], format="%Y-%m-%d")
full_data['RecentTrainingExperience2From_datetime'] = pd.to_datetime\
(full_data['RecentTrainingExperience1From'], format="%Y-%m-%d")
full_data['RecentTrainingExperience3From_datetime'] = pd.to_datetime\
(full_data['RecentTrainingExperience1From'], format="%Y-%m-%d")
```

```python
full_data['RecentTrainingExperience1To_datetime'] = pd.to_datetime\
(full_data['RecentTrainingExperience1To'], format="%Y-%m-%d")
full_data['RecentTrainingExperience2To_datetime'] = pd.to_datetime\
(full_data['RecentTrainingExperience2To'], format="%Y-%m-%d")
full_data['RecentTrainingExperience3To_datetime'] = pd.to_datetime\
(full_data['RecentTrainingExperience3To'], format="%Y-%m-%d")
```

```python
date = ['ApplyDate', 'JobPostingDate', 'RecentTrainingExperience1From',\
                'RecentTrainingExperience2From', 'RecentTrainingExperience3From',\
                'RecentTrainingExperience1To', 'RecentTrainingExperience2To', \
                'RecentTrainingExperience3To']
date_features = ['ApplyDate_datetime', 'JobPostingDate_datetime', 'RecentTrainingExperience1From_datetime',\
                'RecentTrainingExperience2From_datetime', 'RecentTrainingExperience3From_datetime',\
                'RecentTrainingExperience1To_datetime', 'RecentTrainingExperience2To_datetime', \
                'RecentTrainingExperience3To_datetime']
```

```python
date_num_features = []
for fea in date_features:
    full_data[fea + '_year']=full_data[fea].apply(lambda x:x.year)
    full_data[fea + '_month']=full_data[fea].apply(lambda x:x.month)
    full_data[fea + '_day']=full_data[fea].apply(lambda x:x.day)
    full_data[fea + '_dayofweek']=full_data[fea].apply(lambda x:x.dayofweek)
    full_data[fea + '_dayofyear']=full_data[fea].apply(lambda x:x.dayofyear)
    full_data[fea + '_weekofyear']=full_data[fea].apply(lambda x:x.weekofyear)
    date_num_features.append(fea + '_year')
    date_num_features.append(fea + '_month')
    date_num_features.append(fea + '_day')
    date_num_features.append(fea + '_dayofweek')
    date_num_features.append(fea + '_dayofyear')
    date_num_features.append(fea + '_weekofyear')
```

```python
full_data.drop(date_features, axis = 1, inplace = True)
full_data.drop(date, axis = 1, inplace = True)
full_data.shape
```

#### Lable encoding

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
LE_features = []
LE_map = dict()
for fea in cat_features:
    print ('Label Encoding %s' % (fea))
    LE_var = fea+'_le'
    full_data[LE_var] = LE.fit_transform(full_data[fea].astype(str))
    LE_features.append(LE_var)
    LE_map[fea] = LE.classes_
```

#### OneHot encoding

```python
from sklearn.preprocessing import OneHotEncoder
OH = OneHotEncoder(sparse=False)
for fea in LE_features:
    print ('One-hot Encoding %s' % (fea))
    temp = OH.fit_transform(full_data[[fea]])
    temp = pd.DataFrame(temp,columns=[(fea + '_' + str(i)) for i in full_data[fea].value_counts().index])
    temp = temp.set_index(full_data.index.values)
    full_data = pd.concat([full_data,temp],axis=1)
```

# Prepare train, validation and test set

```python
for fea in num_features:
    full_data[fea].fillna((full_data[fea].median()), inplace=True)
```

```python
full_data.drop(cat_features, axis = 1, inplace = True)
```

```python
X_train = full_data.loc[full_data['hired'].notnull()]
y = X_train['hired'].values
X_test = full_data.loc[full_data['hired'].isnull()]
```

```python
X_train.hired.value_counts()
```

### Upsample minority class for balanced data

```python
from sklearn.utils import resample
# Separate majority and minority classes
X_train_majority = X_train[X_train.hired == 0]
X_train_minority = X_train[X_train.hired == 1]
 
# Upsample minority class
df_minority_upsampled = resample(X_train_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=X_train.hired.value_counts()[0],    # to match majority class
                                 random_state=123) # reproducible results
```

```python
X_train_upsampled = pd.concat([X_train_majority, df_minority_upsampled])
y_upsampled = X_train_upsampled['hired'].values
```

```python
X_test = X_test.drop(['hired'], axis=1)
X_train = X_train.drop(['hired'], axis=1)
X_train_upsampled = X_train_upsampled.drop(['hired'], axis=1)
```

```python
X_train.shape, y.shape, X_train_upsampled.shape, y_upsampled.shape,  X_test.shape
```

# Modeling

```python
from sklearn.model_selection import train_test_split
X_split_train, X_val, y_train, y_val = train_test_split(X_train_upsampled, y_upsampled, test_size=0.3, random_state=123)
```

### Logistic Regression benchmark

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_split_train, y_train)
```

```python
y_pred = lg.predict(X_val)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_val, y_pred)  
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
print('precision = ', precision_score(y_val, y_pred))
print('recall = ', recall_score(y_val, y_pred))
print('f1_score = ', f1_score(y_val, y_pred, average='macro'))
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_split_train, y_train)
```

```python
y_pred = rf.predict(X_val)
```

```python
confusion_matrix(y_val, y_pred)  
```

```python
print('precision = ', precision_score(y_val, y_pred))
print('recall = ', recall_score(y_val, y_pred))
print('f1_score = ', f1_score(y_val, y_pred, average='macro'))
```
